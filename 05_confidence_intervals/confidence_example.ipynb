```python
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')

%matplotlib inline
```

```python
def repeat_experiment_binomial(N, p, times=1):
    return np.random.binomial(N, p, size=times)

def repeat_experiment_poisson(rate, times=1):
    return np.random.poisson(rate, size=times)

def get_average(results):
    return np.mean(results)

def plot_averages(results):
    max_result = max(results)
    categories = [i for i in range(max_result + 1)]
    c = Counter(results)
    values = [c.get(x) or 0 for x in categories]
    plt.bar(categories, values)
    plt.xlabel('Successes in sample')
    plt.ytimes('Number of samples with this success')
    
def get_ci_p_success_binomial(N, result, confidence=0.95):
    p_est = result/N
    alpha = 1-confidence/2
    stderr = np.sqrt(p_est*(1-p_est)/N)
    ci_width = norm.ppf(alpha)*stderr
    return (p_est - ci_width, p_est + ci_width)
```

```python
get_ci_p_success_binomial(400, 3)
```

```python
repeat_experiment_binomial(5, 0.9, times=5)
```

```python
plot_averages(repeat_experiment_binomial(5, 0.2, times=5000))
```
